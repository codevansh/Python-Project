```python
def shuffle_and_split(data,test_ratio):

  # this will give random permutations everytime we run this so for that we use np.random.seed(42 or any number) to have the same permutations(reproducibility)
  np.random.seed(42)
  shuffled_indices = np.random.permutation(len(data))
  #  this gives shuffled indices
  print(shuffled_indices)
  test_set_size = int(len(data) * test_ratio) # Cast the result of the multiplication to an integer
  test_indices = shuffled_indices[:test_set_size]
  train_indices = shuffled_indices[test_set_size:]
  return data.iloc[train_indices],data.iloc[test_indices]

  # e.g
  # data = [1,2,3,4,5,6,7,8,9]
  # shuffle_and_split(data, .4)
```

```python
shuffle_and_split(df,0.4)
```

```python
df['income_cut'] = pd.cut(df['median_income'], bins =[0,1.5,3.0,4.5,6.0,np.inf], labels=[1,2,3,4,5])
```

```python
df
```

```python
df['income_cut'].value_counts().sort_index().plot.bar(grid=True)
plt.title("income categories distribution")
plt.xlabel("Income Category")
plt.ylabel("No of Instances")
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df,df['income_cut']):
  strata_train_set = df.loc[train_index]
  strata_test_set = df.loc[test_index]
```

```python
strata_train_set
```

```python
strata_train_set['income_cut'].value_counts().sort_index().plot.bar(grid=True)
plt.title("income categories distribution")
plt.xlabel("Income Category")
plt.ylabel("No of Instances")
plt.show()
```

```python
strata_test_set['income_cut'].value_counts().sort_index().plot.bar(grid=True)
plt.title("income categories distribution")
plt.xlabel("Income Category")
plt.ylabel("No of Instances")
plt.show()
```
