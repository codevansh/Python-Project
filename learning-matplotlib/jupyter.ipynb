```python
# x=[2023,2024,2025,2026]
# x=np.array([2023,2024,2025,2026])
# y=np.array([15,25,30,20])

# plt.plot(x,y)
# plt.show()

x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])

line_style = dict(marker=".",
         markersize=20,
         markerfacecolor="yellow",
         markeredgecolor="red",
         linestyle="solid",
         linewidth=3,
         )


# it needs ** to be used or technically it is said as
# the line_style needs to be unpacked form the dict

plt.plot(x,y1,color="pink",**line_style)
plt.plot(x,y2,**line_style)


# Part-2
plt.title("class size",
          fontsize=20)

plt.xlabel("Year",
           fontsize=20)

plt.ylabel("Students",
           fontsize=20)

plt.xticks(x)
plt.tick_params(axis="both")
                #colors="orange"

plt.show()
```

```python
# scatter graph

x1 = np.array([0,1,1,2,3,4,5,6,7,7,8])  #hrs studied
y1 = np.array([55,60,65,62,68,70,75,78,82,85,87]) #grades scored

x2 = np.array([0,1,2,2,3,4,5,6,7,8,8])  #hrs studied
y2 = np.array([50,58,65,70,72,78,83,88,92,95,97]) #grades scored

plt.scatter(x1,y1,color="blue",
            alpha=0.5,
            s=100,
            label="CLass A")

plt.scatter(x2,y2,color="red",
            alpha=0.5,
            s=100,
            label="CLass B")

plt.title("Hrs studied and Grades Co-relation")
plt.xlabel("Hours Studied")
plt.ylabel("Grades")

plt.legend()
```
